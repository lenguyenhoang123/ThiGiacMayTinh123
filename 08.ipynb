```python
import matplotlib.pyplot as plt
```

```python
# biểu đồ line chart
x=[1,2,3,4,5]
y=[10,17,20,30,40]

x1=[1,2,3,4,6]
y1=[20,18,9,10,25]

plt.plot(x,y,'bo-',label='Dữ liệu')# vẽ đường 1
plt.plot(x1,y1,marker='o',color='r',linestyle='-',label='Dữ liệu 1')#duong 2
plt.legend()#chú giải
plt.xlabel('Trục x')
plt.xlabel('Trục y')
plt.show()
```

```python
#bar chart
categories=['A','B','C','D','E']
values=[20,25,30,35,10]

plt.bar(categories, values, color='g',alpha=0.6)
plt.xlabel('Danh muc')
plt.ylabel('Gia tri')
plt.show()
```

```python
#pie chart
my_labels=['A','B','C','D']
sizes=[15,30,40,20]
my_colors=['red','green','blue','yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%')
plt.title('Bieu do hinh tron')
plt.show()
```

```python
  # Vẽ biểu đồ hộp
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 40, 50, 60, 70]

plt.boxplot(data)
plt.title('Biểu đồ hộp ')  
plt.ylabel('Giá trị')  
plt.show()
```

```python
# Vẽ biểu đồ violin
import matplotlib.pyplot as plt

data = [15, 20, 22, 30, 35, 45, 50, 60, 70]

plt.violinplot(data)  
plt.title('Biểu đồ violin')
plt.ylabel('Giá trị')
plt.show() 
```
