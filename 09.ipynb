```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#DUong dn nh
my_path='C:\\ThiGiacMayTinh1\\img\\anh2.JPG'
#c nh
im=Image.open(my_path)
plt.imshow(im)
```

```python
#chuyn sang xm
im_gray=im.convert('L')
plt.imshow(im_gray,cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#To mt hnh
plt.figure()
plt.hist(im_array.flatten(),bins=128)
plt.title('Biu  histogram')
plt.xlabel('Gi tr cu pixel')
plt.ylabel('S Lng')
plt.show()
```

```python
#lm ni bt
plt.figure()
plt.gray()
plt.contour(im_gray, origin='image')
plt.show()
```
