```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

image_path='C:\\ThiGiacMayTinh1\\img\\anh4.png'
image= Image.open(image_path)
# nh xm
gray_image = image.convert('L')

# Chuyn nh xm v mng
image_a = np.array(gray_image)

#tnh gradient theo hng x(o hm ring theo x) bng mt n Prewitt x
gradient_x= ndimage.convolve(image_a, np.array([[-1, -1, -1],[-1, 0, 1],[-1,0,1]]))
#tnh gradient theo hng y(o hm ring theo y) bng mt n Prewitt y
gradient_y = ndimage.convolve(image_a, np.array([[-1,-1,-1],[0,0,0],[1,1,1]]))
#tnh bin  gradient
gradient_magnitude = np.sqrt(gradient_x**2+gradient_y**2)
#hin th nh
fig, axs = plt.subplots(2,1, figsize=(10, 8))

plt.subplot(2,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title('nh xm gc')

plt.subplot(2,1,2)
plt.imshow(gradient_magnitude,cmap='gray')
plt.title('nh xm gc')
plt.show()
```
