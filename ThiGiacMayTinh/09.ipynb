```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#DUong dẫn ảnh
my_path='C:\\ThiGiacMayTinh1\\img\\anh2.JPG'
#Đọc ảnh
im=Image.open(my_path)
plt.imshow(im)
```

```python
#chuyển sang xám
im_gray=im.convert('L')
plt.imshow(im_gray,cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
#Tạo một hình
plt.figure()
plt.hist(im_array.flatten(),bins=128)
plt.title('Biểu đồ histogram')
plt.xlabel('Giá trị cảu pixel')
plt.ylabel('Số Lượng')
plt.show()
```

```python
#làm nổi bật
plt.figure()
plt.gray()
plt.contour(im_gray, origin='image')
plt.show()
```

```python
#Đường dẫn
my_path='C:\\ThiGiacMayTinh1\\img\\anh3.jpg'
#đọc ảnh và hiển thị với trục 
im=Image.open(my_path).convert('L')
#hiển thị ảnh trên trục
plt.imshow(im)
#Tạo một hình mới
plt.figure()
#Dat mau xam
plt.gray()
#vẽ biểu đồ đường biên
plt.contour(im,origin="image")
#Đảm bảo tỷ lệ trucjx và y bằng nhau
plt.axis('equal')
#Hiển thị ảnh
plt.show()
```
