```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

image_path='C:\\ThiGiacMayTinh1\\img\\anh5.png'
image= Image.open(image_path)
# nh xm
gray_image = image.convert('L')

# Chuyn nh xm v mng
image_array = np.array(gray_image)
#chuyn sang nh nh phn
binary_image=1*(image_array<200)

# Thc hin opening  loi b nhiu v ni cc i tng
opened_image = ndimage.binary_opening(binary_image, structure=np.ones((5,5)))

#nh nhn
labeled_image,num_features= ndimage.label(opened_image)
#To li sublot vi 3 dng v 1 ct
fig, axs= plt.subplots(4, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image,cmap='gray')
plt.title("nh xm")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("nh nh phn")

plt.subplot(4,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("nh sau opening")

plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='gray')
plt.title(f" nh nhn({num_features}i tng)")

plt.show();
```
