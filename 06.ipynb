```python
from PIL import Image
from imgtools import *
```

```python
#Duong dan
my_path='C:\\ThiGiacMayTinh1\\img\\anh2.jpg'
```

```python
#doc anh
img=load_image(my_path)
```

```python
display(img)
```

```python
# chuyển từ ảnh RGB sang ảnh xám
gray_image=img.convert("L")
display(gray_image)
```

```python
#phân RGB thành 3 tầng màu khác nhau
red_band, green_band,blue_band=img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```
